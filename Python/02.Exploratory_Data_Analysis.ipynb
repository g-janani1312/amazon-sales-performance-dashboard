# BUSSINESS KPIs

```python
#Total_Orders
total_orders=df["Order ID"].nunique()
print("Total_Orders:",total_orders)
```

```python
#Total_Revenue
total_revenue=df["Amount"].sum()
print("Total_Revenue:",round(total_revenue,2))
```

```python
#Average_Order_value
average_order_value=df["Amount"].mean()
print("Average_Order_value:",round(average_order_value,2))
```

```python
#Total_Quantity_Sold
quantity_sold=df["Qty"].sum()
print("Total_Quantity_Sold:",quantity_sold)
```

```python
#Total_Categories
total_categories=df["Category"].nunique()
print("Total_Categories:",total_categories)
```

```python
# Total States
total_states = df["ship-state"].nunique()
print("Total States:", total_states)
```

```python
# Total Cities
total_cities = df["ship-city"].nunique()
print("Total Cities:", total_cities)
```

```python
# Total Products (SKU)
total_products = df["SKU"].nunique()
print("Total Products:", total_products)
```

# SALES ANALYSIS

```python
#Sales_by_Category
category_sale=(df.groupby("Category")["Amount"].sum().sort_values(ascending=False))
print(category_sale)
```

```python
plt.figure(figsize=(12,6))
category_sale.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_sale.plot(kind='bar',x='Category',y='Revenue',title='Sales by Category',figsize=(12,6), xlabel='Category',ylabel='Revenue',rot=45, color='Green')
```
